```python
from twilio.rest import Client
import pandas as pd
import time

def build_personalized_message(row):
    """
    Construye el mensaje de voz personalizado.

    Args:
        row: El registro del archivo Excel.

    Returns:
        El mensaje de voz personalizado.
    """
    intro_message = "Hola, somos de Falabella.com y tenemos un mensaje."
    personalized_message = f"{intro_message} ¡Hola {row['sellerName']}! Tu pedido asociado al SKU {row['Sku']} está {row['Status']}. Gracias por tu atención."
    return personalized_message

def send_personalized_voice_calls(sender_number):
    """
    Envía llamadas de voz personalizadas.

    Args:
        sender_number: El número de teléfono que se utilizará como remitente de las llamadas.
    """

    account_sid = "AC32986c07b07c11aa0f3482490c759333"
    auth_token = "251b02843eb9391741767d18811ef620"
    client = Client(account_sid, auth_token)

    # Ruta del archivo Excel
    file_path = "/content/data 1 whatsapp - copia.xlsx"  # Actualiza con tu propia ruta

    try:
        df = pd.read_excel(file_path)
        print(f"Archivo cargado: {file_path}")

        # Filtra los pedidos con el estado "canceled"
        canceled_orders = df[df["Status"] == "canceled"]

        # Itera a través de los pedidos cancelados y envía una llamada a cada destinatario
        for index, row in canceled_orders.iterrows():
            # Construye el mensaje de voz personalizado
            personalized_message = build_personalized_message(row)

            # Obtiene el número de teléfono del destinatario
            phone_number = f"+51{str(row['Numero'])}"

            try:
                # Crea la llamada
                call = client.calls.create(
                    to=phone_number,
                    from_=sender_number,
                    twiml=f'<Response><Say voice="Mia" language="es-MX">{personalized_message}</Say></Response>',
                )

                # Agrega un tiempo de espera entre llamadas para evitar problemas con la velocidad de envío
                time.sleep(10)
                print(f"Llamada realizada a {phone_number} correctamente. SID de la llamada: {call.sid}")

            except Exception as e:
                print(f"Error al enviar llamada a {phone_number}: {str(e)}")

    except Exception as e:
        print(f"Error al cargar el archivo: {str(e)}")

# Solicitar el número del remitente
sender_number = input("Número Twilio (From): ")

# Enviar llamadas de voz
send_personalized_voice_calls(sender_number)
```
