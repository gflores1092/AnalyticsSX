```python
from twilio.rest import Client
import pandas as pd
import time

# Función para enviar SMS
def send_sms(sender_number):
    account_sid = 'AC32986c07b07c11aa0f3482490c759333'
    auth_token = '251b02843eb9391741767d18811ef620'
    client = Client(account_sid, auth_token)

    # Ruta del archivo Excel
    file_path = "/content/data 1 whatsapp - copia.xlsx"  # Actualiza con tu propia ruta

    try:
        df = pd.read_excel(file_path)
        print(f"Archivo cargado: {file_path}")

        # Filtra los registros con el estado "canceled"
        canceled_orders = df[df["Status"] == "canceled"]

        # Itera a través de los pedidos cancelados y envía el mensaje a cada destinatario
        for index, row in canceled_orders.iterrows():
            # Construye el mensaje de SMS
            message = f"¡Hola {row['sellerName']}! Tu pedido asociado al SKU {row['Sku']} está {row['Status']}. Gracias por tu atención."

            # Número de teléfono al que se enviará el mensaje
            phone_number = f"+51{str(row['Numero'])}"  # Agrega el código de país

            try:
                # Envía el mensaje de SMS
                client.messages.create(from_=sender_number, body=message, to=phone_number)

                # Agrega un tiempo de espera entre mensajes para evitar problemas con la velocidad de envío
                time.sleep(10)
                print(f"Mensaje enviado a {phone_number} correctamente.")
            except Exception as e:
                print(f"Error al enviar mensaje a {phone_number}: {str(e)}")

    except Exception as e:
        print(f"Error al cargar el archivo: {str(e)}")

# Solicitar el número del remitente
sender_number = input("Número Twilio (From): ")

# Enviar SMS
send_sms(sender_number)
```
